##How do find similar beer drinkers by using written reviews only?

```
#feature extraction
from sklearn.feature_extraction.text import TfidfVectorizer
df = df.sample(frac=1).reset_index()
vectorizer = TfidfVectorizer(stop_words={'english'})
X = vectorizer.fit_transform(df['review_text'])
```

```
from sklearn.cluster import KMeans
```

```
wss = []
for k in range(1,15):
    kmeans = KMeans(n_clusters=k,init='k-means++')
    kmeans=kmeans.fit(X)
    wss_li = kmeans.inertia_
    wss.append(wss_li)
```

```
cnum = range(1,15)
```

```
best_k =pd.DataFrame({'cluster': cnum,'wss':wss})
best_k
```

```
sns.set_style("darkgrid")
sns.scatterplot(x='cluster',y='wss',data=best_k)
```

```
from sklearn.cluster import KMeans
#Initialize the class object
km1=KMeans(n_clusters=10,  max_iter=500, n_init=10)
```

```
#predict the labels of clusters.
km2 = km1.fit_predict(X)
```

```
df['label'] = km2
```
